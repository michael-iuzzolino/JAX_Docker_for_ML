```python
batch_size = 100
download_dir = '/root/mount/jax/datasets'
ys = []
for handler_key in ['tensorflow']:
    for dataset_name in ['CelebA']:#, 'MNIST', 'CIFAR10', 'CIFAR100', 'CelebA']:

        build_params = {
            "download_dir" : download_dir, 
            "dataset_name" : dataset_name, 
            "batch_size"   : batch_size, 
            "flatten_img"  : False, 
            "onehot_label" : True,
            "image_resize" : 224 if dataset_name == 'CelebA' else None
        }

        pt_data_handler = modules.datasets.BuildDataHandler(handler_key, **build_params)

        for x, y in pt_data_handler('train'):
            ys.append(y)
            break

        print(f"Dataset: {dataset_name}")
        print(f"X: {x.shape}\ny: {y.shape}\n")

        modules.utils.show_samples(pt_data_handler, 5)
```
